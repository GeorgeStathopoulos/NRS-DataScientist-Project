```python
import pandas as pd
import numpy as np
```

```python
db_v01 = pd.read_csv("db_v01.csv", sep = ",")
db_v01.head(5)
```

```python
#Convert strings with comma to floats
db_v01["FPP_to_School_enrollment"] = db_v01["FPP_to_School_enrollment"].apply(lambda x: float(x.replace(",",".")) if isinstance(x,str) else x)
db_v01["FRP_Take_up_percent_"] = db_v01["FRP_Take_up_percent_"].apply(lambda x: float(x.replace(",",".")) if isinstance(x,str) else x)
db_v01["EZ_Pay_Take_Up_Rate"] = db_v01["EZ_Pay_Take_Up_Rate"].apply(lambda x: float(x.replace(",",".")) if isinstance(x,str) else x)
db_v01["FPP_to_PAX"] = db_v01["FPP_to_PAX"].apply(lambda x: float(x.replace(",",".")) if isinstance(x,str) else x)
```
