<font size="6">1.1 Import Data and Integration</font>

```python
import pandas as pd
import numpy as np
```

```python
db_crm = pd.read_csv("crm_model.csv", sep = ",")
db_sales = pd.read_csv("sales_model.csv", sep = ",")
db_finance = pd.read_csv("finance_model.csv", sep = ",")
db_twitter = pd.read_csv("twitter_model.csv", sep = ",")
#db_crm.head(5)
```

```python
db_crm['ID'] = db_crm['ID_CRM'].astype(str).str[1:-1] .str.extract('(\d+)', expand=False).astype(int) #.astype(np.int64)     #pd.to_numeric(df_crm['ID'])  #df_crm['ID'] = df_crm['ID'].astype(int)
db_crm.head(5)
```

```python
db_sales['ID'] = db_sales['ID_SALES'].str.extract('(\d+)', expand=False).astype(int)
db_sales.head(5)
#df_finance.head(5)
#df_twitter.head(5)
```

```python
db_finance['ID'] = db_finance['ID_FINANCE'].str.extract('(\d+)', expand=False).astype(int)
db_finance.head(5)
```

```python
db_merg=pd.merge (  pd.merge(db_crm,db_sales,left_on="ID",right_on="ID",how='outer'), db_finance, how='outer' )   # 'outer', how= 'left', how = 'right'
db_merg.head(5) 
#db_merg.shape
```

  <font size="6">1.2 Sentiment Analysis (Vader) -> to check</font>

```python
# I think that the more simply for sentiment is Vader thah make automatically clean
# We can evaluarte some Topic Model...but is too much complex
db_twitter['ID'] = db_twitter['ID_SALES'].str.extract('(\d+)', expand=False).astype(int)
db_twitter.head(5)
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
text_an = SentimentIntensityAnalyzer()
#import nltk
#nltk.downloader.download('vader_lexicon')
```

```python
db_twitter['scores'] = db_twitter['text'].apply(lambda text: text_an.polarity_scores(text))
db_twitter['compound']=db_twitter['scores'].apply(lambda score_dict: score_dict['compound'])
db_twitter.head(5)
```

```python
db_twitter["num_words"]=db_twitter["text"].apply(lambda x: len(x.split()))
db_twitter.head()
```

```python
db_twitter_gr=db_twitter.groupby("ID").mean()
db_twitter_gr["num_tweets"] = db_twitter.groupby("ID")["text"].count()
db_twitter_gr["compound_min"] = db_twitter.groupby("ID")["compound"].min()
db_twitter_gr["compound_max"] = db_twitter.groupby("ID")["compound"].max()
db_twitter_gr["num_words_max"] =db_twitter.groupby("ID")["num_words"].max()
db_twitter_gr["num_words_min"] = db_twitter.groupby("ID")["num_words"].min()
db_twitter_gr.head(5)
```

```python
db_v01=pd.merge(db_merg,db_twitter_gr,left_on="ID",right_on="ID",how='outer')
db_v01.head(5)# try how ='outer', how= 'left', how = 'right'
db_v01.shape
```
